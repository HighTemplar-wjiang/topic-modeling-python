```python
import os
import numpy as np
import pandas as pd
from collections import OrderedDict
from pprint import pprint

# Figures.
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Topic modeling.
from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import contractions 

# Download the WordNet data needed for lemmatization (commented out as internet access is disabled)
# nltk.download('wordnet')
```

```python
# Figure presets.
preset_figsize = (6, 5)
preset_axislabelsize = 15
preset_axislabelweight = "bold"
preset_ticklabelsize = 15
preset_ticklabelweight = "bold"
```

```python
# Define the file path
file_path = 'sample_data.txt'

# Initialize lists to hold data
participants = []
questions = []
answers = []

# Read the file and process the data
with open(file_path, 'r') as file:
    lines = file.readlines()

    # Iterate through lines and extract the data
    for i in range(0, len(lines), 4):  # Step by 4 (3 lines of data + 1 blank line)
        participants.append(lines[i].strip())
        questions.append(lines[i + 1].strip())
        answers.append(lines[i + 2].strip())

# Create a DataFrame with lowercase column names
df = pd.DataFrame({
    'participant_id': participants,
    'question_id': questions,
    'answer': answers
})

# Display the DataFrame
df.head(5)
```

```python
# Define additional stop words to remove general words like "feel", "scene" and their derivations
additional_stop_words = ['feel', 'feeling', 'felt', 'scene', 'bit', 'little', 'lot']
included_tags = ["NN", "JJ"]
n_top_words = 5

# Initialize the WordNet lemmatizer
lemmatizer = WordNetLemmatizer()


# Function to map POS tag to first character used by WordNetLemmatizer
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # Default to NOUN


# Function to lemmatize a document with POS tagging
def lemmatize_doc_with_pos(doc):
    tokens = nltk.word_tokenize(doc.lower())  # Tokenize and convert to lower case
    pos_tags = nltk.pos_tag(tokens)  # Get POS tags
    pos_tags = list(filter(lambda x: x[1] in included_tags, pos_tags)) # Filter words.
    lemmatized_tokens = [lemmatizer.lemmatize(token, get_wordnet_pos(pos)) for token, pos in pos_tags]
    return ' '.join(lemmatized_tokens)


# Update the function to include lemmatization
def perform_lda_visualization_lemmatized(text_data, n_topics=3, n_top_words=5, additional_stop_words=None, 
                                         word_cloud=False, debug=False, verbose=False,
                                         wordcloud_width=1920, wordcloud_height=1080,
                                         basefontsize=20, basefontweight="bold", 
                                         colors=plt.cm.Blues(np.linspace(1, 0, n_top_words)),
                                         savename="", savetype="pdf", savepath="./"):
    # Function to expand contractions in a text such as didn't
    text_data = [contractions.fix(doc) for doc in text_data]
    
    # Lemmatize the text data
    lemmatized_text_data = [lemmatize_doc_with_pos(doc) for doc in text_data]

    # Debug.
    if debug:
        print(lemmatized_text_data)
    
    # Combine built-in stop words with additional stop words
    stop_words = 'english'
    if additional_stop_words:
        stop_words = set(CountVectorizer(stop_words='english').get_stop_words())
        stop_words.update(additional_stop_words)
    
    # Vectorize the text data
    vectorizer = CountVectorizer(stop_words=list(stop_words))
    dtm = vectorizer.fit_transform(lemmatized_text_data)
    
    # Perform LDA
    lda = LatentDirichletAllocation(n_components=n_topics, random_state=0)
    lda_result = lda.fit_transform(dtm)
    
    # Get the feature names (words)
    feature_names = vectorizer.get_feature_names_out()
    
    # Initialize the plot
    fig, axes = plt.subplots(1, n_topics, figsize=(20, 5), sharex=False, dpi=300)
    axes = axes.flatten()
    
    # Plot the topics
    current_max = 0
    for topic_idx, topic in enumerate(lda.components_):
        top_features_ind = topic.argsort()[:-n_top_words - 1:-1]
        top_features = [feature_names[i] for i in top_features_ind]
        weights = topic[top_features_ind]

        if verbose:
            print(f"Topic {topic_idx+1} and its weights:")
            print(list(zip(top_features, weights)))
            print()
        
        ax = axes[topic_idx]

        # Get colors.
        if type(colors) is list:
            barcolors = colors[topic_idx]
        else:
            barcolors = colors
        
        # Set xlim.
        if np.max(weights) > current_max:
            current_max = np.max(weights)
        ax.barh(top_features, weights, height=0.7, ec="k", color=barcolors)
        ax.set_title(f'Topic {topic_idx + 1}',
                     fontdict={'fontsize': basefontsize, 'fontweight': basefontweight})
        ax.invert_yaxis()
        ticks = ax.get_yticks()  # Get current tick locations
        ax.set_yticks(ticks)  # Explicitly set tick locations
        ax.set_yticklabels(ax.get_yticklabels(), fontsize=basefontsize, fontweight=basefontweight)

    # Set x axis.
    for ax in axes:
        ax.set_xlim([0, np.ceil(current_max)+1])
        ticks = ax.get_xticks()  # Get current tick locations
        ax.set_xticks(ticks)  # Explicitly set tick locations
        ax.set_xticklabels(ax.get_xticklabels(), fontsize=basefontsize, fontweight=basefontweight)
    fig.tight_layout()

    # Word cloud. 
    if word_cloud:
        # Draw word cloud.
        wc = WordCloud(background_color="white", colormap="Set2", 
                       width=wordcloud_width, height=wordcloud_height,
                       stopwords=stop_words).generate(" ".join(lemmatized_text_data))
        fig_wc, ax_wc = plt.subplots(dpi=300)
        ax_wc.axis("off")
        ax_wc.imshow(wc)

        if verbose:
            from operator import itemgetter
            word_counts = WordCloud().process_text(" ".join(lemmatized_text_data))
            word_counts = OrderedDict(sorted(word_counts.items(), key=itemgetter(1)))
            print("Word counts:")
            print(list(word_counts.items()))
            # for (word, count) in word_counts.items():
            #     print(f"{word}: {count}")

    # Save figures.
    if savename:
        fig.savefig(os.path.join(savepath, f"{savename}_topic_modeling.{savetype}"), dpi=300, bbox_inches="tight")
        fig_wc.savefig(os.path.join(savepath, f"{savename}_wordcloud.{savetype}"), dpi=300, bbox_inches="tight")
        
    return lda_result, feature_names, lemmatized_text_data
```

```python
# Topic modeling.
# Settings.
n_topics = 3
n_topwords = 5
figure_path = "../figures"
wordcloud_width = 1920
wordcloud_height = 1080

# For figure color map. 
# List of all sequential colormaps
sequential_colormaps = [
    'Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
    'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu', 'GnBu', 
    'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn', 'viridis', 
    'plasma', 'inferno', 'magma', 'cividis'
]

# Use an OrderedDict to hold color lists for each colormap
color_lists = []

# Generate color lists for each colormap using np.linspace
for cmap_name in sequential_colormaps:
    color_list = mpl.colormaps.get_cmap(cmap_name)(np.linspace(0.5, 0, 5))
    color_list[:, 3] = 0.6  # Set the alpha value for all colors
    color_lists.append(color_list)


# Loop through each unique question_id in the DataFrame
for idx, question_id in enumerate(df['question_id'].unique()):
    # Slice the DataFrame by the current question_id
    question_df = df[df['question_id'] == question_id]
    
    # Combine text.
    # combined_text = filtered_survey_df[filtered_survey_df[filter_column] == scene_name][text_column]
    combined_text = question_df["answer"].fillna("NaN").tolist()
    
    # LDA. 
    lda_result, feature_names, lemmatized_text_data = perform_lda_visualization_lemmatized(
        combined_text, n_topics=n_topics, n_top_words=n_top_words, debug=False, verbose=True,
        additional_stop_words=additional_stop_words, word_cloud=True,
        colors=color_lists[idx], savename=f"{question_id}", savetype="png", savepath=figure_path)

    plt.show()
```
